# Removing Outliers

```python
per25th=df_salaries["salary_in_usd"].quantile(0.25)
print(per25th)
per75th=df_salaries["salary_in_usd"].quantile(0.75)
print(per75th)
```

```python
IQR = per75th - per25th
```

```python
upper =per75th +(IQR*1.5)
lower = per25th - (IQR * 1.5)

print(upper,lower)
```

```python
outli= df_salaries[(df_salaries["salary_in_usd"] > upper) | (df_salaries["salary_in_usd"] < lower)]
```

```python
df_salaries= df_salaries[~df_salaries.index.isin(outli)]
```

```python
df_salaries=df_salaries.drop(outli.index)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder

# 
df = df_salaries 
X = df[['company_size', "employment_type",
        'job_title', 'remote_ratio','experience_level',
        "salary_currency",'employee_residence','company_location']]
y = df['salary_in_usd']  

#   
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X)

#     
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.35, random_state=112)

#  
model = LinearRegression()
model.fit(X_train, y_train)

#  
y_pred = model.predict(X_test)

#  MSE  R
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

#  
print(f"Mean Squared Error: {mse}")
print(f"R Score: {r2}")

#   
new_data = [['L', "FT", "Data Scientist", 50, "SE", 'USD', 'US', 'US']]  
new_data_encoded = encoder.transform(new_data)
predicted_salary = model.predict(new_data_encoded)
print(f'Predicted Salary: {predicted_salary}')
```

```python
from sklearn.linear_model import Ridge, Lasso
model_ridge = Ridge(alpha= 10)
model_ridge.fit(X_train, y_train)
y_pred_ridge = model_ridge.predict(X_test)
print("Ridge Model R:", r2_score(y_test, y_pred_ridge))

model_lasso = Lasso(alpha=10)
model_lasso.fit(X_train, y_train)
y_pred_lasso = model_lasso.predict(X_test)
print("Lasso Model R:", r2_score(y_test, y_pred_lasso))
```
